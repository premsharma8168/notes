## Regular expressions or RegEx
It is a sequence of characters mainly used to find or replace patterns embedded in the text.

## 1. re.match()

The re.match function returns a match object on success and none on failure

```python
import re

#match a word at the beginning of a string
result = re.match('The',r'The cat jumped gracefully onto the windowsill, chasing a beam of sunlight.') #change it with 'The'
print(result)
```

```python
print(result.group())
```

# 2. re.search()

Matches the first occurrence of a pattern in the entire string(and not just at the beginning)

```python
result = re.search('bright',r'The sun is bright. She wore bright colors.')
print(result.group())
```

# 3. re.findall()

It will return all the occurrences of the pattern from the string

```python
result = re.findall('bright',r'The sun is bright. She wore bright colors.')
print(result)
```

## Some patterns

1. /b: returns a match where the specified pattern is at the beginning or at the end of a word.

```python
str = r'The cat jumped gracefully onto the windowsill, chasing a beam of sunlight.'
```

```python
x = re.findall(r"\bjum", str)
print(x)
```

2. /d: returns a match where the string contains digits (numbers from 0-9)

```python
str = "Today's date is 6 August 2025"
```

```python
x = re.findall("\d", str)
print(x)

if (x):
  print("There is a match")
else:
  print("No match")
```

```python
x = re.findall("\d+", str)
print(x)

if (x):
  print("There is a match")
else:
  print("No match")
```

3. \D: returns a match where the string does not contain any digit. It is basically the opposite of \d.

```python
x = re.findall("\D", str)
print(x)

if (x):
  print("There is a match")
else:
  print("No match")
```

```python
x = re.findall("\D+", str)
print(x)

if (x):
  print("There is a match")
else:
  print("No match")
```

# Metacharacters

characters with special meaning

1. (.)  matches any character (except newline character)

```python
str = r'The cat jumped gracefully onto the windowsill, chasing a beam of sunlight.'
```

```python
x = re.findall("ju.", str)
x2 = re.findall("ju...", str)

print(x)
print(x2)
```

2. (^) starts with

It checks whether the string starts with the given pattern or not.

```python
x = re.findall("^The", str)

if (x):
  print("Yes")
else:
  print("No match")
```

3. ($) ends with

It checks whether the string ends with the given pattern or not.

```python
str="Today is a very good weather"
```

```python
x = re.findall("weather$", str)

if (x):
  print("Yes")
else:
  print("No match")
```

4. (*) matches for zero or more occurrences of the pattern to the left of it

```python
str1 = "easy easssy eay ey"
```

```python
x = re.findall("eas*y", str1)
print(x)

if (x):
  print("Yes")
else:
  print("No match")
```

5. (+) matches one or more occurrences of the pattern to the left of it

```python
x = re.findall("eas+y", str1)
print(x)

if (x):
  print("Yes")
else:
  print("No match")
```

6. (?) matches zero or one occurrence of the pattern left to it.

```python
x = re.findall("eas?y",str1)

print(x)

if (x):
  print("Yes")
else:
  print("No match")
```

7. (|) either or

```python
x = re.findall("very|good", str)
print(x)

if (x):
  print("Yes")
else:
  print("No match")
```
